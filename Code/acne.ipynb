```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
img = cv2.imread('../Data/IMG_6509.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img, 'gray')
```

```python
plt.imshow(img_rgb, 'gray')
```

```python
img_lab = cv2.cvtColor(img, cv2.COLOR_BGR2Lab)
plt.imshow(img_lab)
```

```python
L, a, b = cv2.split(img_lab)
plt.imshow(a, 'gray')
```

```python
k = np.ones((5,5), np.float32)/25
img_filter = cv2.filter2D(a, -1, k)
plt.imshow(img_filter, 'gray')
```

```python
ret, img_threshold = cv2.threshold(img_filter, 147, 255, cv2.THRESH_BINARY)
#img_threshold = cv2.adaptiveThreshold(img_filter, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 5, 5)
plt.imshow(img_threshold, 'gray')
```

```python
plt.imshow(img_filter, 'gray')
```

```python
erosion_dilation_kernel = np.ones((5,5), np.uint8)
erosion = cv2.erode(img_threshold, erosion_dilation_kernel, iterations = 1)
plt.imshow(erosion, 'gray')
```

```python
dilation = cv2.dilate(erosion, erosion_dilation_kernel, iterations = 5)
plt.imshow(dilation, 'gray')
```

```python
tmp, contours, hierarchy = cv2.findContours(img_threshold, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
cv2.drawContours(img, contours, -1, (0, 0, 255), 3)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb, 'gray')
```
